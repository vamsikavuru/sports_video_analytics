```
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('/content/df.jpg')
```

```
#cv2.imshow("hello", img)
cv2_imshow(img)
```

```
n_rows, n_colms = img.shape[:2]
```

```
n_rows
```

```
n_colms
```

```
matrix_translation = np.float32([[1,0,60], [0,1,90]])
```

```
matrix_translation
```

```
img_translation = cv2.warpAffine(img,matrix_translation,(n_rows, n_colms))
```

```
cv2_imshow(img_translation)
```

Image Rotation



```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow 
```

```
img = cv2.imread("/content/go.png")
```

```
cv2_imshow(img)
```

```
# ratate of an Image
n_rows, n_columns = img.shape[:2]
```

```
n_rows
```

```
rotation_matrix = cv2.getRotationMatrix2D((n_rows/2, n_columns/2), 45, 1)
```

```
print(rotation_matrix)
```

```
rotation_angle = cv2.warpAffine(img,rotation_matrix,(n_rows,n_columns))
```

```
cv2_imshow(rotation_angle)
```
